````python
# @title Decorators
def dec(fun):
    def wrapper():
        print("Before Calling the decorated function.")
        fun()
        print("After Calling the decorated function.")
    return wrapper

@dec        # Calling decorator using @
def say_hello():
    print("Good Morning!")

say_hello()

@dec
def print_log():
    print("This is log!")

print_log()
````

````python
def dec(fun):
    def wrapper(*args,**kwargs):
        print("Before Calling the decorated function.")
        fun(*args,**kwargs)
        print("After Calling the decorated function.")
    return wrapper

@dec
def print_sum(a,b):
    print(f"{a} + {b} = {a+b}")

print_sum(10,20)
````

````python
def dec(fun):
    def wrapper(*args,**kwargs):
        print("Before Calling the decorated function.")
        fun(*args,**kwargs) # We are not returning or printing the result
        print("After Calling the decorated function.")  # It will print None in the output
    return wrapper

@dec
def get_sum(a,b):
    return a+b

print(get_sum(10,20))
````

````python
def dec(fun):
    def wrapper(*args,**kwargs):
        print("Before Calling the decorated function.")
        result = fun(*args,**kwargs)
        print("result is :",result)     # Printing the result
        print("After Calling the decorated function.")
        return result   # Returning the result
    return wrapper

@dec
def get_sum(a,b):
    return a+b

print(get_sum(10,20))
````

````python
def loop(n):
    def dec(fun):
        def wrapper():
            print("Before Calling the decorated function.")
            for _ in range(n):
                fun()
            print("After Calling the decorated function.")
        return wrapper
    return dec

@loop(3)
def say_hello():
    print("Good Morning!")

say_hello()
````

````python
import traceback
from types import TracebackType
try:
    print("Before error line")
    a = 1/0
    print("After error line")
except ZeroDivisionError as e:
    print(e)
    traceback.print_exc()
except Exception as e:
    print(e)
else:
    print("Inside Else")   # If no error occured this block will execute
finally:
    print("Inside finally")
````

````python
import traceback
# Define Custom Exceptions
class AppError(Exception):
    """Base class for application exceptions"""
    pass

class ValidationError(AppError):
    def __init__(self, message):
        super().__init__(message)

n = input("Enter a number:")
try:
    if not n.isnumeric():
        raise ValidationError("Invalid numeric input from the user.")
    print("You entrered a valid number:",n)
except ValidationError as e:
    print(e)
    traceback.print_exc()
````

````python
# Custom Context Manager using class
class ManagedFile:
    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file

    def __exit__(self,exc_type, exc_val, exc_to):
        if self.file:
            self.file.close()

with ManagedFile('hello.txt', 'w') as f:
    f.write('Hello, World!')
````

````python
from contextlib import contextmanager

@contextmanager
def open_file(name,mode):
    f = open(name,mode)
    try:
        yield f
    finally:
        f.close()

with open_file('text_data.txt','r') as f:
    data = f.read()
    print(data)
````

````python
from os import stat_result
from contextlib import contextmanager

import time

@contextmanager
def timer():
    start = time.time()
    yield
    end = time.time()
    print(f"Elapsed: {end - start:.2f} seconds")

with timer():
    sum([i*i for i in range(1000000)])
````

````python
# Logging to output streamer/file
import logging
logger = logging.getLogger("myLogger")
logger.setLevel(logging.INFO)
logger.propagate = False

# Remove all existing handlers
if logger.hasHandlers():
    logger.handlers.clear()

stream_handler = logging.StreamHandler()
logger.addHandler(stream_handler)

dev_file_handler = logging.FileHandler("app_dev.log")
logger.addHandler(dev_file_handler)

test_file_handler = logging.FileHandler("app_test.log")
logger.addHandler(test_file_handler)

logger.info("Info message passed to the handlers.")
logger.critical("CRITICAL mea=ssage passed to the handlers.")
logger.debug("DEBUG message passed to the handlers.")
````

````python
# Formatter and logging to a File with Rotation
import logging
from logging.handlers import RotatingFileHandler


logger = logging.getLogger("my_logger")
logger.setLevel(logging.INFO)
# Remove all existing handlers
if logger.hasHandlers():
    logger.handlers.clear()


handler = RotatingFileHandler("app.log", maxBytes=200, backupCount=3)
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s - %(lineno)d - %(funcName)s')
handler.setFormatter(formatter)
logger.addHandler(handler)


logger.info("Application started")
logger.warning("Warning log")
logging.shutdown() # ensures the close/release of all the file handlers
````

##### Note: Format parameter fields:
```
Placeholder	Description

%(levelname)s	Log level name (e.g., INFO, DEBUG, WARNING, ERROR, CRITICAL).
%(message)s	The actual message passed to the logger (logger.info(...), etc.).
%(name)s	Name of the logger (my_logger in your case).
%(filename)s	The filename where the log call was made.
%(lineno)d	Line number in the source file.
%(funcName)s	Function name containing the logging call.
%(threadName)s	Name of the current thread.
%(process)d	Process ID.
%(module)s	Module name.
%(pathname)s	Full path of the source file.
